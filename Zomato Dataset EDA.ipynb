```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

##### Observation: 
1. When Rating is 4.5 and above, it is considered as an Excellent rating.
2. When Rating is between 4.0 and 3.4, it is considered as Very Good.
3. When Rating is between 3.5 and 3.9, it is considered as Good.
4. When Rating is between 3.0 and 3.4, it is considered as Average.
5. When Rating is between 2.5 and 2.9, it is considered as Average.
6. When Rating is between 2.0 and 2.4, it is considered as Poor.


```python
sns.barplot(x= 'Aggregate rating', y= 'Rating Count', data = ratings, hue = 'Rating color', palette=['white', 'red', 'orange', 'yellow', 'green', 'green'])
plt.title('Relation between Aggregate Rating and Rating Count')
plt.xlabel('Aggregate Rating')
plt.ylabel('Rating Count')
plt.xticks(rotation=90)
plt.figure(figsize=(10, 6))
plt.show()
```

##### Observations:
1. Not Rated count is the highest, which means many restaurants are not rated.
2. Maximum restaurants have an aggregate rating of 2.5, which is considered average.

```python
# Count Plot 
sns.countplot(x = 'Rating color', data = ratings, palette=['white', 'red', 'orange', 'yellow', 'green', 'green'])
plt.title('Count of Ratings by Color')
plt.show()
```
